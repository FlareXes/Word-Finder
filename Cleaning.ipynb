```python
def clean_and_tokenize_data(filename):
  with open(filename, "r") as file:
    data = file.read()

  object = json.loads(data)
  stop_words = stopwords.words('english')
  word_list = []
  for word in object:
      if word not in stop_words:
          word_list.append(word)

  tokenized_words_list = []
  tokenized_words_length = []
  for i in word_list:
    tokenized_word = list(i)
    tokenized_words_list.append(tokenized_word)
    tokenized_words_length.append(len(tokenized_word))

  preprocessed_data = pd.DataFrame({"tokenized_word": tokenized_words_list, "word length": tokenized_words_length})
  preprocessed_data.to_csv(cleaned_tokenized_word_dict_csv)
```

```python
def clean_and_tokenize_data_csv(filename):
  database = pd.read_csv("unigram_freq.csv")
  stop_words = stopwords.words('english')
  word_list = []

  for word in database['word']:
    if word not in stop_words:
        word_list.append(word)
  
  tokenized_words_list = []
  tokenized_words_length = []
  for i in word_list:
    tokenized_word = list(str(i))
    tokenized_words_list.append(tokenized_word)
    tokenized_words_length.append(len(tokenized_word))

  preprocessed_data = pd.DataFrame({"tokenized_word": tokenized_words_list, "word_length": tokenized_words_length})
  preprocessed_data.to_csv("new_cleaned.csv")
  
```

```python
from nltk.corpus import stopwords
import pandas as pd
import json
import nltk
import numpy as np
import collections
import ast

cleaned_tokenized_word_dict_csv = 'new_cleaned.csv'

def dfLenFilter(userWord):
  userWordLen = len(userWord)
  userWordList = list(userWord)

  database = pd.read_csv(cleaned_tokenized_word_dict_csv)
  filter_1_len = database.loc[database['word_length'] ==  userWordLen, :]
  filter_1_len_list = filter_1_len["tokenized_word"].tolist()
  
  answer = []
  for i in filter_1_len_list:
    i = ast.literal_eval(i)
    filter_2_same_elements = set(i) == set(userWordList)
    if filter_2_same_elements == True:
      possible_word = "".join(i)
      answer.append(possible_word)
  return pd.DataFrame({"word": answer})
```
